```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
```

```python
error = np.random.randn(300)
first_integral = np.cumsum(error)
second_integral = np.cumsum(first_integral)
```

```python
plt.plot(error, 'b-', label='error')
plt.plot(first_integral, 'r-', label='$\int^t e(s) ds$')
#plt.plot(second_integral, 'g-', label='$\int^t \int^s e(s) ds dt$')
plt.xlabel('time')
plt.legend()
```
